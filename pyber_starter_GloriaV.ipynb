```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "city_data.csv"
ride_data_to_load = "ride_data.csv"

# Read the City and Ride Data
city = pd.read_csv(city_data_to_load)
c = city.sort_values('city')

ride = pd.read_csv(ride_data_to_load)
r = ride.sort_values('city')

# Combine the data into a single dataset
final_data = pd.merge(r, c, how = 'outer', on = 'city')

# Display the data table for preview
display(final_data)
```

## Bubble Plot of Ride Sharing Data

```python
# Pre-processing / setting iloc indexer for dataframe
city_d = c.iloc[:,0]
driver_count = c.iloc[:,1]
city_type = c.iloc[:,2]

total_no_of_rides_per_city = []
average_fare = []

total_no_of_rides_per_city = final_data.groupby('city')['city'].count()
    #print(total_no_of_rides_per_city)
    
average_fare = final_data.groupby('city')['fare'].mean()
#print(average_fare)

df1 = pd.DataFrame(list(zip(city_d, total_no_of_rides_per_city, average_fare, driver_count, city_type)), columns = ['city','number of rides','fare','driver count','city type'])
display(df1)


# Obtain the x and y coordinates for each of the three city types

# For urban
x1 = df1.loc[df1['city type'] == 'Urban','number of rides']
y1 = df1.loc[df1['city type'] == 'Urban','fare']

# For suburban
x2 = df1.loc[df1['city type'] == 'Suburban','number of rides']
y2 = df1.loc[df1['city type'] == 'Suburban','fare']

# For rural
x3 = df1.loc[df1['city type'] == 'Rural','number of rides']
y3 = df1.loc[df1['city type'] == 'Rural','fare']


# Build the scatter plots for each city types

plt.scatter(x1, y1, c = 'lightcoral', s = driver_count,label = 'Urban')
plt.scatter(x2, y2, c = 'lightskyblue', s = driver_count, label = 'Suburban')
plt.scatter(x3, y3, c = 'gold', s = driver_count, label = 'Rural')

# Incorporate the other graph properties
plt.title('Pyber Ride Sharing Data(2016)', size = 'x-large')
plt.xlabel('Total Number of Rides(per city)', size = 'large')
plt.ylabel('Average fare ($)', size = 'large')
plt.grid(True)

# Create a legend
plt.legend(loc = 0) 

# Incorporate a text label regarding circle size
plt.text(43, 35, 'Note:\n Circle size correlates with driver count per city')

# Save Figure
plt.savefig('Pyber.png', bbox_inches = 'tight')

plt.show()
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents

total_fare = sum(final_data.iloc[:, 2].values)
    #display(total_fare)

city_type = ['Rural','Suburban','Urban']

total_fare_type = final_data.groupby('type')['fare'].sum()
    #display(total_fare_type)

percent = total_fare_type * 100 / total_fare
    #display(percent)

df2 = pd.DataFrame(list(zip(city_type, total_fare_type, percent)), columns = ['city type','total fare','Fare %'])
display(df2)

# Build Pie Chart
explode = (0, 0, 0.1) #explode the last slice
colors = 'gold', 'lightskyblue', 'lightcoral'

plt.pie(percent, explode = explode, labels = city_type, colors = colors, autopct = '%1.1f%%', shadow = True, startangle = 140)
plt.tight_layout()
plt.title('% Of Total Fare By City Type')

# Save Figure
plt.savefig('PyberFare.png',bbox_inches='tight')

plt.show()
```

```python
# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents

total_rides = sum(df1.iloc[:, 1].values)
    #display(total_rides)

city_type = ['Rural', 'Suburban', 'Urban']
 
total_rides_type = df1.groupby('city type')['number of rides'].sum()
    #display(total_rides_type)

percent = total_rides_type * 100 / total_rides
    #display(percent)

df3 = pd.DataFrame(list(zip(city_type, total_rides_type, percent)),columns=['city type', 'total rides', 'Rides %'])
display(df3)

# Build Pie Chart
explode = (0, 0, 0.1) #explode the last slice
colors = 'gold', 'lightskyblue', 'lightcoral'
plt.pie(percent,explode = explode, labels = city_type, colors = colors, autopct = '%1.1f%%', shadow = True, startangle = 140)
plt.tight_layout()
plt.title('% Of Total Rides By City Type')

# Save Figure
plt.savefig('PyberRides.png', bbox_inches = 'tight')

plt.show()
```

```python
# Show Figure
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Drivers Percents

total_drivers = sum(df1.iloc[:, 3].values)
#display(total_drivers)

city_type = ['Rural', 'Suburban', 'Urban']

total_drivers_type = df1.groupby('city type')['driver count'].sum()
#display(total_drivers_type)

percent = total_drivers_type * 100 / total_drivers
#display(percent)

df4 = pd.DataFrame(list(zip(city_type, total_drivers_type, percent)), columns = ['city type', 'total drivers', 'drivers %'])
display(df4)

# Build Pie Chart
explode = (0, 0, 0.1) #explode the last slice
colors = 'gold', 'lightskyblue', 'lightcoral'
plt.pie(percent, explode = explode, labels = city_type,colors = colors, autopct = '%1.1f%%', shadow = True, startangle = 140)
plt.tight_layout()
plt.title('% Of Total drivers By City Type')

# Save Figure
plt.savefig('PyberDrivers.png', bbox_inches = 'tight')

plt.show()
```
